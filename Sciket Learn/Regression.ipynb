### Line function
* y = wx + b

```python
def equation(x , w, b = 0):
    return w*x + b
```

### Loss Function

```python
def mean_square_error(y_true, y_prediction):
    s = (y_true - y_prediction) ** 2
    return np.sqrt(s.mean())   
```

```python
fig, ax = plt.subplots(1,2,figsize = (10,5))
data.plot.scatter(x = 'Height',y = 'Weight',ax=ax[0])
bs = [-100, -50, 0 , 50 , 100 , 150]
mses = []
for b in bs:
    y_pred = equation(x,2,b)
    mse = mean_square_error(y,y_pred)
    mses.append(mse)
    ax[0].plot(x, y_pred)
    
    
ax[1].plot(bs, mses,'o-')
```

```python
fig, ax = plt.subplots(figsize = (5,5))
ax.scatter(x = data.Height.values, y = data.Weight.values)
bs = [50]
mses = []
for b in bs:
    y_pred = equation(x,2,b)
    mse = mean_square_error(y,y_pred)
    mses.append(mse)
    ax.plot(x, y_pred,'red')
```
